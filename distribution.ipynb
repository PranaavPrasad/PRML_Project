Plotting the distribution of offences that the inmates have done.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'illinois_doc_dataset\csv\sentencing.csv', sep=';')
df.info()
```

```python
df.head()
```

```python
offense = df['offense'].value_counts()
offense
```

```python
cumulatative_varience = np.cumsum(offense.values) / offense.sum()
d_prime = np.argmax(cumulatative_varience > 0.75)
print(f'Number of offenses that make up 75% of the data: {d_prime}')
```

```python
offense[:d_prime].plot(kind='bar', figsize=(12, 6))
plt.title('Offense Distribution - 75% of the data')
plt.xlabel('Offense')
plt.ylabel('Frequency')
plt.show()
```

Plotting all offences

```python
plt.figsize=(12, 6)
plt.bar(offense.index, offense.values)
plt.title('Offense Distribution')
plt.xlabel('Offense')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```
